```python
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel
from dotenv import load_dotenv
from openai.types.responses import ResponseTextDeltaEvent
from openai import AsyncOpenAI
from openai import OpenAI
import os

load_dotenv()
```

```python
client=AsyncOpenAI(api_key=os.getenv('GOOGLE_GEMINI'), base_url="https://generativelanguage.googleapis.com/v1beta/openai/") 

model = OpenAIChatCompletionsModel(
    model="gemini-2.5-flash-lite",
    openai_client=client)
```

```python
instruction1='''You are a sales agents in Soumil corp, a company selling Ai 
teaching tools for teacher. Write a cold email to school pitching our product'''
Sale_agent1=Agent(name='Engaging Sale Agent', instructions=instruction1, model=model)
```

```python
OPENAI_API_KEY=os.getenv('OPENAI_API_KEY')

with trace('Cold Email Agent'):
    result= Runner.run_streamed(Sale_agent1, "Write an email pitching our tool 'Teachify : integrating ai with teaching'")
    print(result)
    async for event in result.stream_events():
        if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):
            print(event.data.delta, end="", flush=True)
    print(result)
```
